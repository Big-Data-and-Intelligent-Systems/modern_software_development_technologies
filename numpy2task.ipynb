```python
import numpy as np
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings("ignore")
```

```python
point_count = 1000
x = np.random.uniform(-10, 10, point_count)
y = np.random.uniform(-10, 10, point_count)
plt.scatter(x, y)
```

```python
def distance(xy1, xy2):
    x1, y1 = xy1
    x2, y2 = xy2
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)

r = []
x1y1 = None
x2y2 = None
```

```python
%%time

for x_coord, y_coord in zip(x, y):
    if x1y1 is None:
        x1y1 = (x_coord, y_coord)  
    elif x2y2 is None:
        x2y2 = (x_coord, y_coord)  
        r.append(distance(x1y1, x2y2))
        x1y1 = None
        x2y2 = None

for i, (x_val, y_val) in enumerate(zip(x, y)):
    print(i, (round(x_val, 2), round(y_val)))

print(f"\nРасстояния между парами: {r}")
```
